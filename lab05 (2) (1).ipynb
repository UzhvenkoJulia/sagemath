Розгляньте групу матриць за множенням: 

$$K = \left\langle \left(\begin{array}{rrr}
-1 & 0 & 0 \\
0 & 1 & \frac{1}{2} \\
0 & 0 & 1
\end{array}\right), 
\left(\begin{array}{rrr}

Дайте відповідь на наступні питання:

2. _Чи є група K комутативною?_
3. _Який вигляд мають елементи групи K?_
4. _Які порядки мають елементи групи K?_ 

__Відповіді__: 
1. нескінченна
2. некомутативна
3.
$$K = \left\{ \left(\begin{array}{rrr}
(-1)^{x} & 0 & y \\
0 & 1 & \frac{1}{2} \cdot x \\
0 & 0 & 1
\end{array}\right) , \, {x, y} \in \mathbb{Z} \right\}$$

$$K = \left\{ \left(\begin{array}{rrr}
(-1)^{x} & 0 & 1 \\
0 & 1 & \frac{1}{2} \cdot x \\
0 & 0 & 1
\end{array}\right) \cdot \left(\begin{array}{rrr}
1 & 0 & z \\
0 & 1 & 0 \\
0 & 0 & 1
\end{array}\right) , \, {x, z} \in \mathbb{Z} \right\}$$

6. нескінченного порядку, (окрім нейтрального - 1)



__Додаткова задача:__
Опишіть підгрупи групи K з точністю до ізоморфізму. 

```python
from random import randint
from sage.all import *

K1 = Matrix([[-1, 0, 0], [0, 1, 1/2], [0, 0, 1]])
K2 = Matrix([[1, 0, 1], [0, 1, 0], [0, 0, 1]])
G = MatrixGroup([K1, K2])

# обчислення порядку групи
def compute_order(G, max_power):
    elements = set()
    for k in range(-max_power, max_power + 1):
        for m in range(-max_power, max_power + 1):
            elements.add(tuple((K1 ** k * K2 ** m).list()))  # зберіг як кортеж
    return len(elements)

def is_abelian(G):
    elements = list(G.gens())
    for i in range(len(elements)):
        for j in range(i, len(elements)):
            if elements[i] * elements[j] != elements[j] * elements[i]:
                return False
    return True

# порядок групи з певним макс для ступенів
max_power = 10  
order = compute_order(G, max_power)
print(f"порядок групи: {order}")
print(f"група є абелевою: {is_abelian(G)}")

while True:
    k = randint(-10, 10)
    m = randint(-10, 10)
    
    if m == 1:
        break
    
    print(K1 ** k * K2 ** m)  

for i in range(100):
    k = randint(-10, 10)
    m = randint(-10, 10)
    print(K1 ** k * K2 ** m)  
```
