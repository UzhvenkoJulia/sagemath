```python
def analyze_group(G, name):
    print(f"група: {name}")
    print(f"к-сть елементів у групі |G| = {G.order()}")
    
    center = G.center()
    print(f"центр групи Z(G): {center}")
    print(f"к-сть елементів у центрі: {center.order()}")
    
    conjugacy_classes = G.conjugacy_classes()
    print(f"к-сть класів спряженості: {len(conjugacy_classes)}")
    print("представники класів спряженості:")
    for cls in conjugacy_classes:
        print(f"  {cls.representative()} (розмір класу: {len(cls)})")
    print("-" * 50)

# 1. Циклічна група C11
C11 = CyclicPermutationGroup(11)
analyze_group(C11, "C11")

# 2. Діедральна група D13
D13 = DihedralGroup(13)
analyze_group(D13, "D13")

# 3. Кватерніонна група Q8
Q8 = QuaternionGroup()
analyze_group(Q8, "Q8")
```

```python
# створ поле GF(7)
F = [0, 1, 2, 3, 4, 5, 6]

def determinant(matrix):
    a, b, c = matrix[0]
    d, e, f = matrix[1]
    g, h, i = matrix[2]
    det = (a * (e * i - f * h) - b * (d * i - f * g) + c * (d * h - e * g)) % 7
    return det

# 3x3 матриці в GF(7) (всі можлив)
def generate_matrices():
    matrices = []
    for a in F:
        for b in F:
            for c in F:
                for d in F:
                    for e in F:
                        for f in F:
                            for g in F:
                                for h in F:
                                    for i in F:
                                        matrix = [[a, b, c], [d, e, f], [g, h, i]]
                                        if determinant(matrix) == 1:  # фільтруємо матриці з визначником 1
                                            matrices.append(matrix)
    return matrices

SL3Z7 = generate_matrices()
print(f"SL(3, Z7) має {len(SL3Z7)} елементів")

# SL(3, Z7) має 5630688 елементів
# 64 conjugacy classes

# рахує дуже довго
# виведемо к-сть класів спряженості та приклади представників
# тут тільки виведемо к-сть класів
# обчислення класів спряженості потребує побудови операції спряження і пошуку рівних елементів
#conjugacy_classes = {}  
#for matrix in SL3Z7:
    #found = False
    #for key in conjugacy_classes:
        #if matrix == conjugacy_classes[key][0]:
            #conjugacy_classes[key].append(matrix)
            #found = True
            #break
    #if not found:
        #conjugacy_classes[len(conjugacy_classes)] = [matrix]

#print(f"SL(3, Z7) має {len(conjugacy_classes)} класів спряженості")
```

```python
# from sage.groups.perm_gps.permgroup_named import AlternatingGroup
# from sage.graphs.graph_generators import graphs
```

```python
from sage.graphs.graph import Graph

def create_graph_with_edges(vertices, edges):
    g = Graph()
    g.add_vertices(range(vertices))  
    for edge in edges:
        g.add_edge(edge[0], edge[1])  
    return g

def verify_group_isomorphism(graph, expected_group_structure, group_label):
    automorphism_group = graph.automorphism_group()
    
    print(f"група автоморфізмів графа {group_label}: {automorphism_group.structure_description()}")
    
    automorphism_structure = automorphism_group.structure_description()
    
    if automorphism_structure == expected_group_structure:
        print(f"група автоморфізмів графа {group_label} ізоморфна з {expected_group_structure}: True")
    else:
        print(f"група автоморфізмів графа {group_label} ізоморфна з {expected_group_structure}: False")
        
# граф для A2 (циклічний граф з 2 вершинами)
vertices_A2 = 2
edges_A2 = [(0, 1)]
graph_A2 = create_graph_with_edges(vertices_A2, edges_A2)
verify_group_isomorphism(graph_A2, 'C2', "A2") 

# граф для A3 (циклічний граф з 3 вершинами)
vertices_A3 = 3
edges_A3 = [(0, 1), (1, 2), (2, 0)]
graph_A3 = create_graph_with_edges(vertices_A3, edges_A3)
verify_group_isomorphism(graph_A3, 'S3', "A3") 

# граф для A4 (граф з 4 вершинами)
vertices_A4 = 4
edges_A4 = [(0, 1), (1, 2), (2, 3), (3, 0), (0, 2), (1, 3)]
graph_A4 = create_graph_with_edges(vertices_A4, edges_A4)
verify_group_isomorphism(graph_A4, 'S4', "A4") 

# граф для A5 (граф з 5 вершинами)
vertices_A5 = 5
edges_A5 = [(0, 1), (1, 2), (2, 3), (3, 4), (4, 0)]
graph_A5 = create_graph_with_edges(vertices_A5, edges_A5)
verify_group_isomorphism(graph_A5, 'D5', "A5")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4), (4, 5), (3, 5)])

pos = nx.planar_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', font_weight='bold', node_size=700)
plt.title("планарне розташування графа")
plt.show()

automorphisms = list(nx.algorithms.isomorphism.GraphMatcher(G, G).isomorphisms_iter())
print(f"к-сть автоморфізмів: {len(automorphisms)}")

# дія групи автоморфізмів на граф

orbits = []
for automorphism in automorphisms:
    orbit = set()
    for vertex in G.nodes():
        orbit.add(automorphism[vertex])
    orbits.append(sorted(list(orbit)))

print("орбіти групової дії:")

for orbit in orbits:
    print(orbit)

vertex = 1  
stabilizer = []
for automorphism in automorphisms:
    if automorphism[vertex] == vertex:
        stabilizer.append(automorphism)

print(f"\nстабілізатор вершини {vertex}: {len(stabilizer)}")

stabilizer_orbits = []
for stabilizer_element in stabilizer:
    orbit = set()
    for v in G.nodes():
        orbit.add(stabilizer_element[v])
    stabilizer_orbits.append(sorted(list(orbit)))

print("орбіти стабілізатора:")
for orbit in stabilizer_orbits:
    print(orbit)
```
