```python
from sage.all import *

D8 = DihedralGroup(8)
Z4 = D8.subgroup(['(1,3,5,7)(2,4,6,8)'])

normal_check = True
for g in D8:
    left_conjugates = [g * z * g**-1 for z in Z4] 
    right_conjugates = [g * z * g**1 for z in Z4]  
    if sorted(left_conjugates) != sorted(right_conjugates): 
        normal_check = False
        break

print("is Z4 a normal subgroup of D8: ", normal_check)
```

```python
from sage.all import *

# розв'язок завдання

D8 = DihedralGroup(8)

# створюємо підгрупу Z4 з пермутаціями (1,3,5,7)(2,4,6,8)

Z4 = D8.subgroup([D8('(1,3,5,7)(2,4,6,8)')])

# порядок підгрупи Z4
print("order of Z4: ", Z4.order())

right_cosets = D8.cosets(Z4, side='right')
left_cosets = D8.cosets(Z4, side='left')

print("right cosets of D8 by Z4: ")
for coset in right_cosets:
    print(coset)

print("left cosets of D8 by Z4: ")
for coset in left_cosets:
    print(coset)

# нормальність підгрупи Z4
sorted_right_cosets = sorted([sorted(coset) for coset in right_cosets])
sorted_left_cosets = sorted([sorted(coset) for coset in left_cosets])
condition_for_normality = sorted_right_cosets == sorted_left_cosets
print("is Z4 a normal subgroup of D8? : ", condition_for_normality)
```

## Завдання 2 

Переберіть всі підгрупи знакозмінної групи $A_5$. Перевірте всі на нормальність. Порівняйте час роботи з методом `.is_simple` для групи $A_5$. 

```python
import time

def check_normality(group, subgroup):
    right_cosets = group.cosets(subgroup, side='right')
    left_cosets = group.cosets(subgroup, side='left')
    sorted_right_cosets = sorted([sorted(coset) for coset in right_cosets])
    sorted_left_cosets = sorted([sorted(coset) for coset in left_cosets])
    return sorted_right_cosets == sorted_left_cosets

A5 = AlternatingGroup(5)

begin_time = time.time()

subgroups = A5.subgroups()
normal_subgroups = [check_normality(A5, subgroup) for subgroup in subgroups]

if normal_subgroups.count(True) == 1:
    print("висновок стосовно A5: A5 є простою групою")
else:
    print("висновок стосовно A5: A5 не є простою групою")

end_time = time.time()

for i in range(len(subgroups)):
    print(f"чи є підгрупа, що генерується {subgroups[i]}, нормальною в A5:", normal_subgroups[i])

print("цей метод порівняння за допомогою косетів зайняв", end_time-begin_time, "секунд")

# перевірка простоти групи за допомогою вбудованої функції
begin_time = time.time()
print(f"чи є A5 простим: {A5.is_simple()}")
end_time = time.time()

print("цей метод з використанням вбудованої функції зайняв", end_time-begin_time, "секунд")
```

### Гіпотеза:

Я вважаю, що для діедральних груп $D_n$, де $n$ — непарне число, порядок нормальних підгруп цієї групи визначається дільниками числа $n$, а також порядком самої групи. Це пояснюється тим, що в таких групах наявні нормальні підгрупи, що відповідають кожному дільнику $n$, а також сама група, що є нормальною підгрупою свого порядку. Це дозволяє стверджувати, що для непарних $n$ нормальні підгрупи мають порядок, який є або дільником $n$, або самим числом $n$, що визначає порядок групи.

Коли ж $n$ парне, картина змінюється. Крім нормальних підгруп, порядок яких є дільниками числа $n$, з'являються ще дві додаткові нормальні підгрупи, які мають порядок рівний самому числу $n$. Ці підгрупи з’являються завдяки особливостям симетрій, що виникають у групах з парним порядком. Враховуючи це, можна припустити, що для групи $D_{470448}$ кількість нормальних підгруп буде складатися з 93 елементів. Це число включає 90 підгруп для всіх дільників числа 470448, одну підгрупу, яка відповідає всій групі, та дві підгрупи, що мають порядок 470448 і пов’язані з підвищеною симетрією групи.

У будь-якій діедральній групі є нормальна циклічна абелева підгрупа, породжена елементом $r$, який відповідає за обертання. В нашому випадку, ця підгрупа буде ізоморфною групі $\mathbb{Z}_{470448}$, а всі інші нормальні підгрупи будуть ізоморфними циклічним групам $\mathbb{Z}_d$, де $d$ є дільником числа 470448. Це дозволяє збудувати багатий набір нормальних підгруп, які пов’язані з різними властивостями обертальних симетрій.

Додатково, дві нормальні підгрупи, які з’являються для парних $n$, є наслідком особливості структури групи. Вони є ізоморфними діедральній групі меншого порядку, вдвічі меншого, тобто $D_{\frac{n}{2}}$. Це дозволяє розглядати групу як результат поділу на дві частини, де кожна частина відповідає за симетрії з певною кількістю вершин.

Крім того, для парних значень $n$, що описують симетрії правильних многокутників з $2k$ вершинами, можливо спростити структуру, усунувши симетрії, що відповідають певній половині вершин, тим самим зменшуючи складність аналізу і дозволяючи зосередитися на симетріях, що пов'язані з іншою частиною вершин. Це дозволяє ще глибше зрозуміти внутрішню структуру груп і, відповідно, скоротити кількість обчислень для розв’язання задач, пов'язаних з діедральними групами.

```python
from sympy import isprime
import math

# знаходження всіх дільників числа
def divisors(n):
    divs = []
    for i in range(1, int(math.sqrt(n)) + 1):
        if n % i == 0:
            divs.append(i)
            if i != n // i:
                divs.append(n // i)
    return divs

# побудова списку порядків нормальних підгруп для діедральних груп
def normal_subgroups_orders(n):
    divs = divisors(n)
    # якщо n парне, додаємо ще дві підгрупи порядку n
    if n % 2 == 0:
        divs.append(n)  # підгрупа для всієї групи
        divs.append(n // 2)  # для порядку 470448, припускаючи додаткові симетрії
    return divs

# перебір діедральних груп Dn для n від 3 до 100
orders_by_n = {}
for n in range(3, 101):
    orders_by_n[n] = normal_subgroups_orders(n)

for n, orders in orders_by_n.items():
    print(f"D_{n}: {orders}")

# прогноз для D_470448
# прогнозуємо на основі гіпотези
n_large = 470448
predicted_normal_subgroups = 90 + 1 + 2
print(f"прогнозоване число нормальних підгруп для D_{n_large}: {predicted_normal_subgroups}")
```

На сайті https://brauer.maths.qmul.ac.uk/Atlas/v3/spor/HS в секції `Representations of HS : Number of Points 100` знайдіть у форматі GAP генератори $G = \langle a, b \rangle$. Задайте їх як елементи групи $S_{100}$ і породіть ними підгрупу $G$. Це так звана Higman-Sims група. Виведіть її порядок і знайдіть нормальні підгрупи. Зробіть висновки. 

```python
from sage.all import *

# Representations of HS : Number of Points 100
# G = <a,b>
# завантаження генераторів групи Higman-Sims із файлів
# GAP
# Higman-Sims

with open('HSG1-p100B0.g1', 'r') as file1:
    gen1_raw = file1.read().strip()  # обрізаю зайві пробіли та нові рядки
    # очищаю рядок від маркерів, зайвих пробілів та символів
    gen1_str = gen1_raw.replace('a11 :=', '').replace(';', '').replace('\n', '').replace(' ', '').strip()
    gen1_perm = Permutation(gen1_str)  # об'єкт пермутації 

with open('HSG2-p100B0.g2', 'r') as file2:
    gen2_raw = file2.read().strip()  
    gen2_str = gen2_raw.replace('a21 :=', '').replace(';', '').replace('\n', '').replace(' ', '').strip()
    gen2_perm = Permutation(gen2_str) 

# S_100
SymGroup100 = SymmetricGroup(100)

# породжую підгрупу Higman-Sims групи, використовуючи два генератори
HigmanSimsGroup = SymGroup100.subgroup([gen1_perm, gen2_perm])

# порядок групи 
print("order of the Higman-Sims group:", HigmanSimsGroup.order())
# Для групи, що породжена генераторами, цей порядок дає нам інформацію про загальну кількість елементів у групі

# всі нормальні підгрупи Higman-Sims групи
print("normal subgroups of the Higman-Sims group:\n")
for normal_subgroup in HigmanSimsGroup.normal_subgroups():
    print(normal_subgroup, end='\n\n')
    # може бути важливо для розуміння її внутрішньої структури та класів подібності

print("is the Higman-Sims group simple:", HigmanSimsGroup.is_simple())
#  дозволяє визначити, чи можна поділити групу на менші нормальні підгрупи. 
# Якщо група проста, це означає, що вона не має нетривіальних нормальних підгруп, 
# що може бути важливим аспектом при вивченні її властивостей. 
# Для групи Higman-Sims можна очікувати, що вона не є простою, оскільки це одна з найбільших простих груп, але вона має складну структуру з підгрупами.

# виводжу генератори для перевірки
print("generator 1:", gen1_perm)
print("generator 2:", gen2_perm)
# Перевірка генераторів дає змогу переконатися, що ми правильно їх імпортували і що вони дійсно породжують підгрупу Higman-Sims. 
# Це також дозволяє підтвердити, що група побудована відповідно до визначення.
```
