```python
# ваш код тут 

# усі перестановки 100 елем, к-сть елем = n!

from math import factorial, gcd

# обчислення НСК
def calculate_lcm(values):
    def lcm(a, b):
        return a * b // gcd(a, b)

    result = 1
    for number in values:
        result = lcm(result, number)
    return result

# генерація цілих частин
def generate_partitions(num):
    return Partitions(num).list()

# підрахунок к-сті перестановок за даною частиною
def calculate_permutations(partition):
    total_permutations = factorial(sum(partition))
    counts = {}
    for length in partition:
        counts[length] = counts.get(length, 0) + 1
    
    divisor = 1
    for length, count in counts.items():
        divisor *= (length ** count) * factorial(count)
    
    return total_permutations // divisor

# обчислення к-стей перестановок у Sn
def compute_sn(n):
    all_partitions = generate_partitions(n)
    order_counts = {}
    
    for partition in all_partitions:
        permutations = calculate_permutations(partition)
        cycle_order = calculate_lcm(partition)
        
        if cycle_order not in order_counts:
            order_counts[cycle_order] = 0
        order_counts[cycle_order] += permutations
    
    return order_counts

"""умова завдання містить неймовірно великі значення чисел;
довго перебирала, виводила, гралася з ними;
на сворення файлів йде дуже багато часу;
сам код повинен неодмінно працювати коректно, але розумію, що вірити от так от на словах не можна"""

n_value = 5 
output_result = compute_sn(n_value)

with open(f"orders_Sn{n_value}.txt", "w") as file_output:
    for order, count in output_result.items():
        file_output.write(f"order: {order}, count: {count} \n")

print("saved")
```

```python
from math import factorial, gcd

# містить парні перестановки S100 -> розмір = n!|2

def compute_lcm(numbers):
    def lcm(a, b):
        return a * b // gcd(a, b)

    cumulative_lcm = 1
    for number in numbers:
        cumulative_lcm = lcm(cumulative_lcm, number)
    return cumulative_lcm

def generate_partitions(num):
    return Partitions(num).list()

def calculate_permutations(partition):
    total_count = factorial(sum(partition))
    frequency = {}
    
    for element in partition:
        frequency[element] = frequency.get(element, 0) + 1
    
    divisor = 1
    for value, count in frequency.items():
        divisor *= (value ** count) * factorial(count)
    
    return total_count // divisor

def compute_sn_permutations(n):
    all_partitions = generate_partitions(n)
    order_count_dict = {}
    
    for partition in all_partitions:
        perm_count = calculate_permutations(partition)
        cycle_order = compute_lcm(partition)
        
        if cycle_order not in order_count_dict:
            order_count_dict[cycle_order] = 0
        order_count_dict[cycle_order] += perm_count
    
    return order_count_dict

def is_even_partition(partition):
    even_count = sum(1 for element in partition if element % 2 == 0)
    return even_count % 2 == 0

# обчислення парних перестановок у An
def compute_an_permutations(n):
    all_partitions = generate_partitions(n)
    order_count_dict = {}
    
    for partition in all_partitions:
        if is_even_partition(partition):
            perm_count = calculate_permutations(partition)
            cycle_order = compute_lcm(partition)
            
            if cycle_order not in order_count_dict:
                order_count_dict[cycle_order] = 0
            order_count_dict[cycle_order] += perm_count
            
    return order_count_dict

n_value = 5 # така ж саме історія з цим An, така як з Sn
output_result = compute_an_permutations(n_value)

with open(f"orders_An{n_value}.txt", "w") as file_output:
    for order, count in output_result.items():
        file_output.write(f"order: {order}, count: {count} \n")

print("saved")
```

```python
# ваш код тут

# n <= 10^6
# У симетричній групі Sn порядок елемента відповідає НСК довжин циклів у його розкладанні на неперетинні цикли

from sage.all import *
import time
import os

# НСК для списку чисел
def LCM_list(lst):  # найменше кратне для списку чисел
    if not lst:
        return 1
    lcm_value = lst[0]
    for num in lst[1:]:
        lcm_value = lcm(lcm_value, num)
    return lcm_value

# перевірка існування елемента порядку k в S_n
def exists_element_of_order_k(k, n):
    if factorial(n) % k != 0:  
        print(f"n! = {factorial(n)} не ділиться на {k} - неможливе існування елемента порядку {k}")
        return False, None
    
    partitions = Partitions(n)  # всі розбиття числа n
    for partition in partitions:
        lcm = LCM_list(partition)  
        print(f"перевіряємо розбиття {partition}, НСК = {lcm}")
        
        if lcm != k: 
            continue
        else:
            return True, partition  
    
    return False, None

def task_3(n, k):
    # існує елемент порядку k у S_n?
    start_time = time.time()
    
    result_filename = f"result_task3_lab03_n{n}_k{k}.txt"
    
    try:
        with open(result_filename, "w") as f:
            # Перевіряємо наявність елемента порядку k в S_n
            exists, partition = exists_element_of_order_k(k, n)
            
            if exists:
                f.write(f"у групі S_{n} існує елемент порядку {k} з типом циклу: {partition}\n")
                print(f"у групі S_{n} існує елемент порядку {k} з типом циклу: {partition}")
            else:
                f.write(f"у групі S_{n} немає елемента порядку {k}\n")
                print(f"у групі S_{n} немає елемента порядку {k}")
        
        print(f"файл з результатами створено: {os.path.abspath(result_filename)}")
    except IOError as e:
        print(f"помилка при створенні або записі файлу: {e}")
    
    print(f"час виконання: {time.time() - start_time:.2f} секунд")

# запуск функції для перевірки
task_3(5, 3)  # значення n і k
```
